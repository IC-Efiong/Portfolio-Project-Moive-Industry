IMPORTING LIBRARIES

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #To adjust the configuration of the plot we will create


#Read in the data
df = pd.read_csv(r'C:\Users\hp\Downloads\movie industry\movies.csv')
```

```python
df.head()
```

CHECKING FOR MISSING DATA

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
df
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,(pct_missing*100)))
```

```python
df
```

CHECKING, SORTING AND CHANGING DATA TYPES

```python
df.dtypes
```

```python
df['yearcorrect'] = df['released'].astype(str).str[:4]
df
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['yearcorrect'] = df['yearcorrect'].astype('int64')
```

```python
df
```

```python
df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.drop_duplicates()
```

```python
df = df.drop(['year'],1)
df
```

PLOTTING OF GRAPHS

```python
## Scatter plot of budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Scatter plot of Budget vs Gross Earnings')
plt.xlabel('Budget(in millions)')
plt.ylabel('Gross Earnings(in millions)')
plt.show()
```

```python
df.head()
```

```python
 ## Regression plot of budget vs gross using seaborn
    
sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

CHECKING CORRELATIONS

```python
df.corr(method='pearson') #different types of correlation, pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix ')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df.head()
```

```python
df_count = df

for col_name in df_count.columns:
    if(df_count[col_name].dtype == 'object'):
        df_count[col_name] = df_count[col_name].astype('category')
        df_count[col_name] = df_count[col_name].cat.codes
        
df_count.head()
```

```python
correlation_matrix = df_count.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix ')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
